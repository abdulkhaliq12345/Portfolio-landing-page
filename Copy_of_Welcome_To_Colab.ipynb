```
import pandas as pd
# Dataset Load Karo
df = pd.read_csv("/content/internship assignment.csv")   # Ya jahan bhi file upload ki hai

print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 Rows:")
print(df.head())
```

```
print("=== ASSIGNMENT 1: Basic Data Analysis ===\n")

# First 5 rows
print("1. First 5 Rows:")
print(df.head())
print("\n" + "="*60)

# Average Age
print("2. Average Age:", round(df['Age'].mean(), 2))

# Highest Glucose wala patient
print("\n3. Patient with Highest Glucose:")
print(df[df['Glucose'] == df['Glucose'].max()][['Pregnancies', 'Glucose', 'Age', 'Outcome']])

print("\n4. Patients with Glucose > 140:")
print(df[df['Glucose'] > 140][['Glucose', 'BMI', 'Age', 'Outcome']].head(10))
```
